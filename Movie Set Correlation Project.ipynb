```python
# Import Libraries
import pandas as pd
import numpy as np #mathematical operations on arrays
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read data
df = pd.read_csv(r'C:\Users\tamse\OneDrive - University of Waterloo\Waterloo Course\Other\Data Analyst Portfolio\Part 4\movies.csv')
```

```python
# Look at data

df.head()
```

```python
# Check Missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types

df.dtypes
```

```python
df = df.dropna() #removed missing data
```

```python
# Removed .0 from budget, gross and votes

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')

df.head()
```

```python
# Create correct year
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Sort
df.sort_values(by=['gross'], inplace = False, ascending=False)
```

```python
# Look at all the data
pd.set_option('display.max_rows', None)
```

```python
# Remove duplicates
df['company'].sort_values(ascending=False)
```

```python
# Scatter plot

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Bugest VS Gross Earnings')
plt.xlabel('Budget')
plt.xlabel('Gross Earnings')
plt.show()
```

```python
# Reg plot

sns.regplot(x='budget',y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Correlation between gross and budget - theres a high correlation

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
